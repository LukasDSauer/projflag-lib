# Eruption flow on flag triple - Movie

There is two different styles of defining the eruption flow. One leaves the points $p_i$ fixed, wheras the other one leaves the points $q_i$ fixed. First, we leave $q_i$ fixed.

```python
fcomplex = construct_from_ratio(1, FlagComplexSvg())

d = draw.Drawing(400, 400,origin=(-200,-200))
# White background for the pictures
drawcirc(d,[0,0],4)
# Render size for the pictures (resolution in terms of image width)
d.setRenderSize(w=1200)

fcomplex.create_triangulation()
#fcomplex.visualize(d, with_inner_triangles = False)
#d
```

We can choose whether we want to show the inner triangles $u_i$ and the helper lines for the construction of the inner triangles.

```python
import copy

triangle = fcomplex.triangles[0]

for r in range(1, 90):
    fcomplex.erupt_triangle(t=-0.2, triangle=triangle, transformation_style="Q")
    
    d1 = copy.deepcopy(d)

    fcomplex.visualize(d1, with_inner_triangles=True, with_helper_lines=True)

    d1.savePng("./temp/efq_" + str(r) + ".png")
```

```python
from moviepy.editor import ImageSequenceClip

images_list = ["temp/efq_"+str(r)+".png" for r in range(1,90)]
animation = ImageSequenceClip(images_list, fps=25)
animation.write_videofile("ef3_style_q.mp4")
#animation.write_gif("eruption_flow_style_q.gif")
animation.ipython_display(fps=25, loop=True, autoplay=True)
```
