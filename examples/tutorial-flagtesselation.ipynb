```python
from flagcomplexSvg import FlagComplexSvg, FlagTesselatorSvg
from flagcomplex.AutoComplex import generate_random
import drawSvg as draw
from flagcomplexSvg.DrawingUtility import drawcirc
```

# Tesselation of a tuple of flags

A tuple of flags defines a pair of nested polygons. By reflecting the inner polygon along its sides, we obtain a convex set which is tesselated by the reflected polygons.

First, we generate a random complex consisting of a couple of flags.

```python
#Generate a complex consisting of 5 flags.
fcomplex = generate_random(5, FlagComplexSvg())
fcomplex.create_triangulation()
fcomplex.draw_complex()
ftesselator = FlagTesselatorSvg(fcomplex, steps = 4)
```

We generate a fitting canvas (the `draw` object `d`). However, we don't need to specify a canvas, the tesselator can also generate one automatically.

Afterwards, we calculate the tesselated convex set. In theory, we could do infinitely many steps for the reflections. Practically, we are limited by our computer's specifications. Here, we will stop after `steps=...`.

```python
d = draw.Drawing(400, 400,origin=(-200,-200))
# White background for the pictures
drawcirc(d,[0,0],4)
# Render size for the pictures (resolution in terms of image width)
d.setRenderSize(w=1200)
 
d = ftesselator.print_image(old = d, with_tiles = True)
d
```

We don't need to show the tiles or the initial polygon.
