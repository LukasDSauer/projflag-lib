# Tesselation of a tuple of flags

A tuple of flags defines a pair of nested polygons. By reflecting the inner polygon along its sides, we obtain a convex set which is tesselated by the reflected polygons.

```python
#Generate a complex consisting of 5 flags.
fcomplex = generate_random(5)
fcomplex.create_triangulation()
fcomplex.draw_complex()
```

We generate a fitting canvas and calculate the tesselated convex set. Note that we could do infinitely many steps for the reflections. Here, we will stop after `steps=`.

```python
d = draw.Drawing(400, 400,origin=(-200,-200))
# White background for the pictures
drawcirc(d,[0,0],4)
# Render size for the pictures (resolution in terms of image width)
d.setRenderSize(w=1200)

ftesselator = FlagTesselator(fcomplex, steps = 4)
ftesselator.print_image(old = d, with_tiles = True)
fcomplex.visualize(d)
```
